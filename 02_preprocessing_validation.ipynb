```python
import pandas as pd
from src.features.tfidf_features import TfidfFeatureExtractor

df = pd.read_csv("../data/processed/cleaned_resumes.csv")

texts = df["cleaned_text"].head(100).tolist()

tfidf = TfidfFeatureExtractor()
X = tfidf.fit_transform(texts)

X.shape
```

```python
resume_text = texts[0]

job_description = """
Looking for a data analyst with experience in Python, SQL,
machine learning, pandas, and data visualization.
"""

resume_vec = tfidf.transform([resume_text])
job_vec = tfidf.transform([job_description])

similarity_score = tfidf.compute_similarity(resume_vec, job_vec)
similarity_score
```

```python
df = pd.read_csv("../data/processed/cleaned_resumes.csv")

# pick a resume from IT / Engineering
it_resume = (
    df[df["Category"].str.contains("Information", case=False, na=False)]
    ["cleaned_text"]
    .sample(1, random_state=42)
    .values[0]
)


job_description = """
Looking for a data analyst with experience in Python, SQL,
machine learning, pandas, data analysis, statistics, and visualization.
"""

resume_vec = tfidf.transform([it_resume])
job_vec = tfidf.transform([job_description])

tfidf.compute_similarity(resume_vec, job_vec)
```
